```python
filename = "file.txt"
file = open(filename, "r")
raw = file.read()

def my_concordance(text, match_search):
    count = 0
    for word in text.split():
        if word == match_search:
            count = count + 1
    return count

my_concordance(raw, "coding")
```

```python
def lexical_diversity(text):
    return len(set(text)) / len(text) * 100

def percentage(count, total):
    return 100 * count / total
    
print(lexical_diversity(text4))
print(percentage(text2.count("love"), len(text2)))
```
