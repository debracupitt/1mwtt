**☼ Compare the lexical diversity scores for humor and romance fiction in 1.1. Which genre is more lexically diverse?**

```python
def lexical_diversity(text):
    return len(set(text)) / len(text) * 100

def percentage(count, total):
    return 100 * count / total

humor_text = brown.words(categories='humor')
romance_text = brown.words(categories="romance")
adventure_text = brown.words(categories="adventure")
belles_lettres_text = brown.words(categories="belles_lettres")
editorial_text = brown.words(categories="editorial")
fiction_text = brown.words(categories="fiction")
government_text = brown.words(categories="government")
hobbies_text = brown.words(categories="hobbies")
learned_text = brown.words(categories="learned")
lore_text = brown.words(categories="lore")
mystery_text = brown.words(categories="mystery")
news_text = brown.words(categories="news")
religion_text = brown.words(categories="religion")
reviews_text = brown.words(categories="reviews")
science_fiction_text = brown.words(categories="science_fiction")
categories = [["CATEGORY", "LEXICAL DIVERSITY"],
              ["", ""],
              ["Humor", str(lexical_diversity(humor_text))],
              ["Romance", str(lexical_diversity(romance_text))],
              ["Adventure", str(lexical_diversity(adventure_text))],
              ["Belles Lettres", str(lexical_diversity(belles_lettres_text))],
              ["Editorial", str(lexical_diversity(editorial_text))],
              ["Fiction", str(lexical_diversity(fiction_text))],
              ["Government", str(lexical_diversity(government_text))],
              ["Hobbies", str(lexical_diversity(hobbies_text))],
              ["Learned", str(lexical_diversity(learned_text))],
              ["Lore", str(lexical_diversity(lore_text))],
              ["Mystery", str(lexical_diversity(mystery_text))],
              ["News", str(lexical_diversity(news_text))],
              ["Religion", str(lexical_diversity(religion_text))],
              ["Reviews", str(lexical_diversity(reviews_text))],
              ["Science-Fiction", str(lexical_diversity(science_fiction_text))]
             ]
```

```python
def print_contents(contents):
    print("Table of Lexical Diversity")
    i = 0
    for c in contents:
        print(contents[i][0].ljust(25) + contents[i][1])
        i = i + 1

print_contents(categories)
```

#### The most lexically diverse category is 'Humor'. The least is 'Learned'.
